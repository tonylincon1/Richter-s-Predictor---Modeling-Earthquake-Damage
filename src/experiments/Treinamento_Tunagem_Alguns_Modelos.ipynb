# Preparação

```python
df_train_values = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Richter's Predictor: Modeling Earthquake Damage/data/train_values.csv")
```

```python
df_train__label_values = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Richter's Predictor: Modeling Earthquake Damage/data/train_labels.csv")
```

```python
df_train_values.head()
```

```python
df_train__label_values.head()
```

```python
print(df_train_values.shape)
print(df_train__label_values.shape)
```

```python
df_train = pd.merge(df_train_values,df_train__label_values,how='inner',on='building_id')
```

```python
print(df_train_values.shape)
print(df_train__label_values.shape)
print(df_train.shape)
```

```python
df_train.damage_grade.value_counts()
```

```python
target = []

for n in df_train.damage_grade:
    if n == 1:
        target.append(0)
    elif n == 2:
        target.append(1)
    elif n == 3:
        target.append(2)
```

```python
df_train["damage_grade"] = target
```

```python
df_train.damage_grade.value_counts()
```

```python
df_teste = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Richter's Predictor: Modeling Earthquake Damage/data/test_values.csv")
```

```python
df_teste.head()
```

# Treinamentos

```python
clf1 = setup(data = df_train, target = 'damage_grade', session_id = 786, silent = True, use_gpu=True)
```

```python
%%time
best_model = compare_models()
```

```python
%%time

dt1 = create_model('lightgbm')
dt2 = create_model('rf')
dt3 = create_model('et')
dt4 = create_model('gbc')
dt5 = create_model('ada')
dt6 = create_model('dt')
dt7 = create_model('ridge')
dt8 = create_model('lda')
dt9 = create_model('nb')
dt10 = create_model('qda')
```

```python
%%time

tuned_dt1 = tune_model(dt1, optimize = 'F1')
```

```python
tuned_dt1
```

```python
predict_tuned_dt1 = predict_model(tuned_dt1, data=df_teste).Label

sub_csv = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Richter's Predictor: Modeling Earthquake Damage/data/submission_format.csv")
sub_csv["damage_grade"] = predict_tuned_dt1
sub_csv.to_csv("/content/drive/My Drive/Colab Notebooks/Richter's Predictor: Modeling Earthquake Damage/output/submit/predict_tuned_dt1.csv", index=False)
```
